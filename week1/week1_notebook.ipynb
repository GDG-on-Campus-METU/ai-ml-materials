## Data types

**Numeric Types**

*   `int`: Represents integers (whole numbers) without decimals, e.g., `10`, `-5`

*   `float`: Represents floating-point numbers (numbers with decimals), e.g., `3.14`, `-2.5`







````python
my_int = 10
my_float = 3.14
print(type(my_int))
print(type(my_float))
````

**Text Type**

*   `str`: Represents strings of characters, enclosed in single or double quotes, e.g., `"Hello"`, `'Python'`.

````python
my_str = "Hello"
print(type(my_str))
````

*   `tuple` An ordered, immutable collection of items, e.g., `(1, 2, "GDSC")`

````python
my_list = [1, 2, "GDSC"]
my_tuple = (1, 2, "GDSC")
print(type(my_list))
print(type(my_tuple))
````

## Operators

**Arithmetic Operators:** Perform basic mathematical operations
*   `+` (Addition)
*   `-` (Substraction)
*   `*` (Multiplication)
*   `/` (Division)
*   `//` (Floor division)
*   `%` (Modulus)
*   `**` (Exponentiation)

````python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
````

**Assignment Operators:** Assign values to variables

*   `=` (Simple assignment)
*   `+=` (Add and assign)
*   `-=` (Substract and assign
*   `*=`, `/=`, `//=`, `%=`, `**=` (Similar for other arithmetic operations)

````python
a = 10
a += 5
print(a)
````

**Comparison Operators:** Compare values and return Boolean results

*   `==` (Equal to)
*   `!=` (Not equal to)
*   `>` (Greater than)
*   `<` (Less than)
*   `>=` (Greater than or equal to)
*   `<=` (Less than or equal to)   

````python
a = 10
b = 5
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
````

**Logical Operators:** Combine Boolean expressions

*   `and` (True if both operands are True)
*   `or` (True if at least one operand is True)
*   `not` (Inverts the truth value)

````python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
````

**Membership Operators:** Test if a value is present in a sequence

*   `in` (Returns True if value is found)
*   `not in` (Returns True if value is not found)

````python
a = [1, 2, 3]
print(1 in a)
print(4 not in a)
````

 **Identity Operators:** Compare the memory locations of objects

*   `is` (Returns True if both operands refer to the same object)
*   `is not` (Returns True if both operands do not refer to the same object)

````python
a = [1, 2, 3]
b = a
print(a is b)
print(a is not b)
````

## Lists

In Python, a list is a versatile and commonly used data structure that stores an ordered collection of items. These items can be of different data types, including numbers, strings, or even other lists.

**Key Features:**

*   **Ordered:** Items in a list maintain the order in which they were added
*   **Mutable:** You can modify a list after it's created  (add, remove, or change items)
*   **Indexed:** Each item in a list is assigned a unique index (starting from 0) that you can use to access it

**Common List Operations:**

*   `append(item)`: Adds an item to the end of the list.
*   `insert(index, item)`: Inserts an item at a specific index.
*   `remove(item)`: Removes the first occurrence of a specific item.
*   `pop(index)`: Removes and returns the item at the specified index (or the last item if no index is given).
*   `len(list)`: Returns the number of items in the list.
*   `sort()`: Sorts the items in the list (in ascending order by default).
*   `reverse()`: Reverses the order of items in the list.

````python
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)
````

````python
my_list = [1, 2, 3]
my_list.insert(1, 5)
print(my_list)
````

````python
my_list = [1, 2, 3]
my_list.remove(2)
print(my_list)
````

````python
my_list = [1, 2, 3]
my_list.pop(1)
print(my_list)
````

````python
my_list = [1, 2, 3]
print(len(my_list))
````

````python
my_list = [3, 1, 2]
my_list.sort()
print(my_list)
````

````python
my_list = [1, 2, 3]
my_list.reverse()
print(my_list)
````

## Functions

**Defining a function:**

*   `def` **keyword:** Signals the start of a function definition.
*   **Function name:** A descriptive name for the function (e.g., `greet`).
*   **Parameters (optional):** Input values enclosed in parentheses.
*   **Docstring (optional):** A brief description of the function's purpose.
*   **Function body:** Indented code block containing the function's logic.

````python
def greet(name):
  """This function greets the person passed in as a parameter."""
  print("Hello,", name + "!")
````

**Calling a function:**

*   Use the function name followed by parentheses.
*   If the function has parameters, provide values inside the parentheses.

````python
greet(name="Alice")
````

**Returning values:**

*   `return` **statement:** Sends a value back to the caller of the function.

````python
def add(x, y):
  """This function adds two numbers."""
  return x + y

result = add(x=5, y=3)
print(result)
````

## Loops

`for` **loop:**

* Iterates over a sequence (like a list, tuple, string, or range) and executes the code block for each item.

````python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
  print(fruit)
````

````python
for i in range(5):
  print(i)
````

`while` **loop:**

*   Executes a code block as long as a condition remains True.

````python
count = 0
while count < 5:
  print(count)
  count += 1
````

`break` **statement:**

*   The break statement is used to immediately terminate the current loop (either for or while) and resume execution at the next statement after the loop.

````python
for i in range(10):
  if i == 5:
    break  # Exit the loop when i equals 5
  print(i)
````

## Classes

In Python, classes act as blueprints for creating objects (instances). They define a set of attributes (data) and methods (functions) that describe the behavior of the objects.

**Defining a class:**

*   `class` **keyword:** Indicates the start of a class definition.
*   **Class name:** Conventionally starts with a capital letter (e.g., `Dog`).
*   `__init__` **method:** A special method (constructor) that initializes the object's attributes when it's created.
*   `self` **parameter:** Refers to the instance of the class itself.
*   **Attributes:** Variables that store data associated with the object (e.g., `name`, `breed`).
*   **Methods:** Functions that define the object's behavior (e.g., `bark`).

````python
class Dog:
  def __init__(self, name, breed):
    self.name = name
    self.breed = breed

  def bark(self):
    print("Hav!")
````

**Creating Objects (Instances):**

*   **Object creation:** Call the class name like a function, passing any required arguments to` __init__`.
*   **Accessing attributes and methods:** Use dot notation (e.g., `my_dog.name`, `my_dog.bark()`).

````python
my_dog = Dog(name="Paşa", breed="Beagle")
print(my_dog.name)
print(my_dog.breed)
my_dog.bark()
````
