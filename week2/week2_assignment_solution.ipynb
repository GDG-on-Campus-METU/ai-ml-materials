## Q1 - NumPy Arrays

1.   Create a numpy array from `arr_list` and assign it to variable `arr`.
2.   Create a numpy array consists of ones and assign it to variable `ones`. Shape of the array must be `(3,3)`.
3.   Create a numpy array consists of intagers from `0` to `9` (including `0`, not including `9`) and assign it to variable `int_arr`.
4.   Change the shape of `int_arr` to `(3,3)`.
5.   Assign first element of `arr` to variable `first_element`.
6.   Assign element wise multiplication of `arr` and `int_arr` to variable `ewmul_arr`.
7.   Assign matrix multiplication of `arr` and `int_arr` to variable `mmul_arr`.

```python
arr_list = [[1,2,3],[4,5,6],[7,8,9]]

#1. Creating an array from a list
arr = np.array(arr_list)

#2. Creating an array consists of ones
ones = np.ones((3,3))

#3. Creating an array consists of integers from 0 to 10
int_arr = np.arange(9)

#4. Changing the shape of an array
int_arr = int_arr.reshape((3,3))

#5. First element of an array
first_element = arr[0,0]

#6. Element-wise multiplication of two arrays
ewmul_arr = arr * int_arr

#7. Matrix multiplication of two arrays
mmul_arr = arr @ int_arr
```

```python
# THIS BLOCK IS FOR GRADING
# DO NOT MODIFY THIS BLOCK

q1_answer = [str(arr), str(ones), str(int_arr), str(first_element), str(ewmul_arr), str(mmul_arr)]
```

## Q2 - Pandas DataFremes

1.   Create a pandas dataframe from `df_dict` and assign it to variable `df`.
2.   Fill missing value with `10` and assign it to variable `filled_df`.
3.   Fill missing values using interpolation and assign it to variable `in_filled_df`.
4.   Convert `in_filled_df` to a numpy array and assign it to variable `np_array`.

```python
df_dict = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, np.nan, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

#1. Creating a dataframe from a dictionary
df = pd.DataFrame(df_dict)

#2. Filling missing values
filled_df = df.fillna(10)

#3. Filling missing values using interpolation
in_filled_df = df.interpolate()

#4. Converting a dataframe to a numpy array
np_array = in_filled_df.to_numpy()
```

```python
# THIS BLOCK IS FOR GRADING
# DO NOT MODIFY THIS BLOCK

q2_answer = [str(df), str(filled_df), str(in_filled_df), str(np_array)]
```
