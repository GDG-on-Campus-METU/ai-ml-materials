# 02. Neural networks


---



A single neuron is fundamentally limited in its computational power. It can only perform a simple linear operation. While this might suffice for very basic problems, it falls short when dealing with complex, real-world data.

Neural networks, on the other hand, are composed of multiple layers of interconnected neurons. This architecture allows them to:

*   **Learn complex patterns:** By stacking layers, neural networks can capture intricate relationships between inputs and outputs that a single neuron cannot.
*   **Handle non-linearity:** Real-world data often exhibits non-linear patterns. Neural networks, equipped with activation functions, can effectively model these complexities.
*   **Achieve hierarchical feature learning:** Deeper networks can learn increasingly abstract features from the data, building upon lower-level representations.
*   **Improve generalization:** With more parameters, neural networks have a greater capacity to fit complex data without overfitting.

## Import libraries

We will use NumPy, Pandas, Matplotlib and TensorFlow in almost every notebook.

Additional libraries may be used in some notebooks.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
print(tf.__version__)
```

## Prepare data

```python
x = np.arange(0, 10, 0.5)
y = x * 2 + 1
```

```python
x_tensor = tf.constant(x, dtype=tf.float32)
y_tensor = tf.constant(y, dtype=tf.float32)
```

## Define and compile the model

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True, show_layer_activations=True, show_layer_names=True, dpi=100)
```

```python
model.compile(
    optimizer = tf.keras.optimizers.SGD(),
    loss = tf.keras.losses.MeanSquaredError(),
    metrics = [tf.keras.metrics.MeanAbsoluteError()]
)
```

## Train the model

```python
model.fit(
    x = x_tensor,
    y = y_tensor,
    epochs = 500
)
```

## Evaluate the model

```python
model.evaluate(x_tensor, y_tensor)
```

```python
test_value = tf.constant([30], dtype=tf.float32)
model.predict(test_value)
```

## Save the model

```python
model.save('model.h5')
```

## Load the model

```python
loaded_model = tf.keras.models.load_model('model.h5')
```

```python
test_value = tf.constant([30], dtype=tf.float32)
loaded_model.predict(test_value)
```
