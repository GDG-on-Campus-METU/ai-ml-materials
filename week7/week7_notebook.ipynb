# 02. Variance and bias

---

**Variance:**

* **Definition:** Variance refers to the sensitivity of the model to fluctuations in the training data.
* **High Variance:** A model with high variance tends to overfit the training data, meaning it learns the noise in the training set and performs well on the training data but poorly on unseen data.

**Bias:**

* **Definition:** Bias is the error introduced by approximating a real-world problem, which may be very complicated, by a much simpler model.
* **High Bias:** A model with high bias tends to underfit the training data, meaning it fails to capture the underlying patterns and performs poorly on both training and test data.

**The Bias-Variance Tradeoff:**

In machine learning, the goal is to find a model with low bias and low variance. However, there's often a tradeoff:

* **Increasing model complexity** (e.g., adding more features or using a more flexible model) tends to decrease bias but increase variance.
* **Decreasing model complexity** tends to increase bias but decrease variance.

Finding the right balance between bias and variance is key to building a model that generalizes well to new data.

# Reducing Variance (Adressing Overfitting)

### Simplify the model

 If your model is overly complex, consider reducing the number of layers, neurons, or parameters. This can help prevent overfitting.

### Feature Selection

* **Correlation Analysis:** Identify and remove highly correlated features.
* **Mutual Information:** Select features that provide the most information about the target variable.
* **Transformations:** Transformations in data science are techniques to modify data for better analysis. They include scaling, normalization, handling missing values, outlier detection, feature engineering, and dimensionality reduction.

### Regularization

* **L2 (Ridge):** Shrinks weights towards zero, but doesn't force them to be exactly zero.
* **L1 (Lasso):** Can force some weights to become exactly zero, effectively performing feature selection.
* **Elastic Net:** Combines L1 and L2 penalties.

```python
tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))
tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l1(0.01))
tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l1_l2(0.01))
```

### Early Stopping

Stop training the neural network when the performance on a validation set starts to decrease, preventing it from overfitting the training data.

```python
tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
```

### Dropout

Randomly deactivates a proportion of neurons during each training iteration, preventing complex co-adaptations between neurons and forcing the network to learn more robust features.

```python
tf.keras.layers.Dropout(0.2)
```

### Data Augmentation

Artificially increase the size and diversity of your training data by applying transformations like rotations, flips, crops, or adding noise to existing samples. This helps the model generalize better.

```python
data_aug_layers = tf.keras.models.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomHeight(0.2),
    tf.keras.layers.RandomWidth(0.2),
    tf.keras.layers.RandomZoom(0.2)
])
```

### Get more data

Providing your model with more diverse examples helps it generalize better and reduces its reliance on specific instances in the training set.

# Reducing Bias (Adressing Underfitting)

### Increase Model Complexity

If your model is too simple, it might not capture the underlying patterns in the data. Try increasing the number of layers and neurons.

### Feature Selection

* **Correlation Analysis:** Identify and remove highly correlated features.
* **Mutual Information:** Select features that provide the most information about the target variable.
* **Transformations:** Transformations in data science are techniques to modify data for better analysis. They include scaling, normalization, handling missing values, outlier detection, feature engineering, and dimensionality reduction.

### Decrease or Remove Regularization

If you have too much regularization, it can excessively constrain the model and lead to underfitting. Try reducing the regularization strength (e.g., the C parameter in logistic regression) or removing it altogether.
